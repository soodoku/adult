### Web Desktop
* For private_domains visited on Windows and OSX devices
* Aggregated to each consecutive change in private_domain
* realityMine_web_desktop_2022-06-01_2022-06-30.csv

```python
df_web_desktop = pd.read_csv("../../adult-data/output/realityMine_web_desktop_2022-06-01_2022-06-30.csv")
print(f"{len(df_web_desktop):,}")
df_web_desktop.head()
```

```python
df_web_desktop.caseid.nunique()
```

```python
adult_cats = [cat for cat in df_web_desktop.dropna(subset="category").category.unique() if "Adult" in cat]
print(len(adult_cats))
adult_cats
```

Some potential false positives (e.g., urbandictionary.com, victoriassecret.com, 4chan.org are not pornsites)

```python
df_web_desktop[df_web_desktop.category.str.contains("Adult", na=False)].private_domain.unique()[:25]
```

```python
for ix, adult_cat in enumerate(adult_cats):
    print(f"{1+ix}: {adult_cat}")
    print('=================')
    _private_domains = df_web_desktop.query(f"category=='{adult_cat}'").private_domain.unique()
    print(_private_domains[:15], "\n")
```

```python
del df_web_desktop
gc.collect()
```
